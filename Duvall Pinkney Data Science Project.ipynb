```python
def dataAnalysis(df):
    for row in df:
        time = df["TIME"]
        location = row[2]
        c = time.find(":")
        hour = int(time[:c])
        if (hour >= 0) & (hour < 1):
            print(df[row])
                
        elif (hour >= 0) & (hour < 1):    
            print(df[row])
                
    minutes = hour / 60        
    plt.plot(minutes, len(row), color='green', marker='o', linestyle='solid')
    plt.title("Collision per hour in 10458")
    plt.axis([0,59,0,24])
    plt.xlabel("Hours")
    plt.ylabel("# of Collisions")
    #plt.show
    return minutes,len(row)
        
#        
```

```python
minutes, y = dataAnalysis(new_df2)

plt.plot(minutes, y, color='green', marker='o', linestyle='solid')
plt.title("Collision per hour in 10458")
plt.axis([0,59,0,24])
plt.xlabel("Hours")
plt.ylabel("# of Collisions")
plt.show
```
